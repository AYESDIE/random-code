```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```
x = np.arange(0, 10)
```

```
x
```

```
y = x ** 2
```

```
plt.plot(x, y, 'r--')
plt.xlim(0, 8)
plt.ylim(0, 80)
plt.title("Y = sq(X)")
plt.xlabel('X')
plt.ylabel('Y')
```
