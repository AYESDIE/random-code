# TF Graphs

```
import tensorflow as tf
```

```
n1 = tf.constant(1)
```

```
n2 = tf.constant(2)
```

```
n3 = n1 + n2
```

```
with tf.Session() as sess:
  result = sess.run(n3)
```

```
result
```
